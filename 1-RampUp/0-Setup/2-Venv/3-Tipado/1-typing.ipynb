&nbsp;&nbsp;&nbsp;&nbsp; En esta clase vamos a hablar de tipado en Python. C칩mo sabr치s para este punto, Python es un lenguaje de tipado din치mico. ## 1.2. Type Hinting


&nbsp;&nbsp;&nbsp;&nbsp; Ahora, 쯖칩mo hago para que se entienda que es lo que la funci칩n espera?. Para lograr eso vamos a utilizar **Type hints**. Si prestaste atenci칩n a la imagen anterior ya habr치s podido ver uno. El formato es `':'` seguido del tipo que esperamos sea nuestra variable (en nuestro caso el `'bool'`, `'= False'` es el valor por defecto).

&nbsp;&nbsp;&nbsp;&nbsp; Los type hints fueron introducidos en el [PEP 484](https://peps.python.org/pep-0484/) y si bien las anotaciones no eran nada nuevo para Python, lo que estableci칩 fue el standard de c칩mo indicar tipos para que se puede hacer un an치lisis est치tico de tipos; asi como, por ejemplo, introducir la libreria standard **Typing** que nos ayuda a espeficar tipos.

&nbsp;&nbsp;&nbsp;&nbsp; Dicho esto, modifiquemos el c칩digo para que sea m치s interpretable:


````python
def multi_call(num: int, alt: bool = False) -> int:
    if alt:
        return num * 5
    return num * - 2

b = multi_call(23)
````

&nbsp;&nbsp;&nbsp;&nbsp; Si solo quisieramos aceptar enteros tendr칤amos que agregar el control a mano:

&nbsp;&nbsp;&nbsp;&nbsp; Obviamente no siempre es el caso. Si nosotros no somos los usuarios deber칤amos hacer chequeos similares al implementeado en la funci칩n anterior. Por ejemplo, si lo que estamos desarrollando es una librer칤a; tendriamos que chequear que el usuario este llamando a la funci칩n de la API con los datos correctos para que no rompa m치s abajo en el stack o mismo si estamos haciendo una API REST en donde los pedidos a un endpoint puede ser tan o m치s creativos que nuestra imaginaci칩n.

&nbsp;&nbsp;&nbsp;&nbsp; Lo que **mypy** busca hacer es tratar de combinar los beneficios de tener un tipado est치tico en el contexto de un lenguaje din치mico. Para eso controla que los tipos que nosotros declaramos como hints coincidan con los de las variables con las que m치s adelante se llaman. Por ejemplo miremos el archivo `classes.py` en donde declaramos dos clases equivalentes `A` y `B`, una de la manera cl치sica de crear clases en Python y la otra con ayuda de la librer칤a estandar `dataclasses` (muy recomendada para evitar codear una y otra vez `__init__`).

````python
%%writefile classes.py

from dataclasses import dataclass


class A:
    def __init__(self, a: int, b: str, c: float):
        self.a = a
        self.b = b
        self.c = c


ia = A(a=1, b=1, c="asdad")


@dataclass
class B:
    a: int
    b: str
    c: float


ib = B(a=1, b=1, c="asdwqe")

print(f"ia: {ia}")
print(f"ib: {ib}")
````

## 1.4. Pydantic

&nbsp;&nbsp;&nbsp;&nbsp; Ahora bien. Como veniamos viendo, type hinting **NO** hace validaci칩n de tipos. Si queremos validar tipos tenemos que chequearlo durante runtime. Para esto podemos usar el viejo y querido `assert` como en el ejemplo anterior o bien, si estamos declarando tipos podemos hacerlo con `Pydantic`.

````python
!pip install pydantic
````

&nbsp;&nbsp;&nbsp;&nbsp; 쮺칩mo funciona? Nada mejor que un buen ejemplo autocontenido:

````python
from datetime import datetime
from typing import List, Optional
from pydantic import BaseModel

class User(BaseModel):
    id: int
    name: str = 'Juan Maniglia'
    signup_ts: Optional[datetime] = None
    friends: List[int] = []

external_data = {'id': 123, 'signup_ts': '2024-06-01 12:22', 'friends': [1, 2, 3, 4]}
user = User(**external_data)
print(user.id)
print(repr(user.signup_ts))
print(user.friends)
print(user.model_dump())
````

&nbsp;&nbsp;&nbsp;&nbsp; Como podemos ver, **pydantic** no solo valida que los tipos de los argumentos con los que estamos llamando sino que, en caso de que no sea v치lido, trata de castearlo hac칤a el especificado en la declaraci칩n. Y digo trata porque no hace magia:

````python
from pydantic import ValidationError

external_data = {'id': 'Pedrito el Corrupto', 'signup_ts': 'ayer', 'friends': ["asd"]}
try:
    user = User(**external_data)
except ValidationError as e:
    print(e.json())
````

&nbsp;&nbsp;&nbsp;&nbsp; En este caso, no existe una forma sencilla o l칩gica de obtener un entero a partir de el string `"Pedrito el Corrupto"` por lo que falla en la validaci칩n al igual que en el resto de los campos. Esto es lo que pasa por ejemplo, si teniendo un endpoint de una API REST construida con [FastAPI](https://fastapi.tiangolo.com/) (que utiliza **pydantic** tras bambalinas) es llamado con un request sin el formato esperado.

&nbsp;&nbsp;&nbsp;&nbsp; Esto est치 genial! 쯇or qu칠 no usarlo siempre? Bueno, por un lado, el concepto que hablamos anteriormente con respecto a cuando deberiamos validar data vs validar llamadas; y por otro, imperceptible al ojo, es menos performante que por ejemplo usar **dataclasses** (que no hace validaci칩n de tipos).


&nbsp;&nbsp;&nbsp;&nbsp; Algunos de los otros features interesantes de **pydantic** incluyen:
* Validators: Podemos validar, adem치s del tipo, cualquier propiedad que se nos ocurran durante la instanciaci칩n de una clase.

````python
from pydantic import BaseModel, ValidationError, field_validator


class UserModel(BaseModel):
    name: str
    username: str

    @field_validator('name')
    def name_must_contain_carlitos(cls, v):
        if 'pedrito' not in v.lower():
            raise ValueError('only pedrito allowed!')
        return v.title()


user = UserModel(
    name='Pedrito el incorruptible',
    username='pedri123',
)
print(user)


try:
    UserModel(
        name='Otro Nombre',
        username='jamr1989',
    )
except ValidationError as e:
    print(e)
````

* Agregar restricciones a nivel de clase. Es decir para todos los argumentos.

````python
from pydantic import BaseModel, ValidationError


class Model(BaseModel):
    a: str
    b: str


    class Config:
        max_anystr_length = 10
        error_msg_templates = {
            'value_error.any_str.max_length': 'max_length:{limit_value}',
        }


try:
    Model(a='x' * 5, b='y' * 15)
except ValidationError as e:
    print(e)
````

* Creaci칩n de schemas en formato JSON a partir de los modelos.

````python
from pydantic import BaseModel, Field
import json

class Gods(BaseModel):
    strengh_level: int
    size: float = None

class Pedrito(Gods):
    name: str = "The one and only"


class Universe(BaseModel):
    """
    This is the description of the main model
    """

    gods: Gods = Field(...)
    pedrito: Pedrito = Field(None, alias='Pedrito')
    snap: int = Field(
        42,
        title='The Snap',
        description='this is the value of snap',
        gt=30,
        lt=50,
    )

    class Config:
        title = 'The universe'


print(json.dumps(Universe.model_json_schema(), indent=2))
````
