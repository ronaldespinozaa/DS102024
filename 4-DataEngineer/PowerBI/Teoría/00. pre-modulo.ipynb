## Instalación y configuración de PowerBI Desktop
- **Instalación**
1. Accedemos a [https://www.microsoft.com/es-es/power-platform/products/power-bi/](https://www.microsoft.com/es-es/power-platform/products/power-bi/).
2. En la sección **Productos**, seleccionamos **Power BI** y luego **Desktop**. Esta versión, recordamos, es totalmente gratuita.
3. En lugar de seleccionar **Descargar gratis** directamente, vamos a **Consulte opciones de descarga e idiomas disponibles**.
4. Se abrirá una nueva pestaña. Allí podemos seleccionar el idioma de preferencia (Español, en nuestro caso).
5. Antes de descargar, es recomendable evaluar los requisitos del sistema un poco más abajo. Se requiere un sistema operativo Windows 10 o superior (si tenemos ordenador Mac o Chromebook, la descarga no es compatible lamentablemente), así como Internet Explorer/Edge 10 o superior. El programa está disponible para plataformas de 32 bits y 64 bits. Para verificar esto en las propiedades de nuestro equipo, vamos al Panel de Control > Sistema (o **Este Equipo**) > Información y observamos el tipo de sistema (64 o 32 bits).
6. Ahora sí, procedemos con la descarga. Elegimos de las dos opciones, la que sea compatible con nuestro sistema. Hacemos clic en **Siguiente** y el programa debería tardar pocos segundos en descargarse en nuestros ordenadores. Si la descarga no comienza automáticamente, el sitio nos ofrece un botón para descarga manual.
7. Una vez descargado el instalador, lo abrimos y ejecutamos. Pasaremos por algunos pasos como la elección de idioma, los términos y condiciones de licencia y la elección de la ubicación en la que almacenar el programa.
8. Una vez finalizada la instalación, hacemos clic en **Finalizar** y automáticamente se debería abrir Power BI Desktop en nuestro ordenador.
